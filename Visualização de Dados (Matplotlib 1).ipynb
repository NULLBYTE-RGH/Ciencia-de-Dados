```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = sns.load_dataset('fmri')
df.head()
```

```python
df = df.sort_values(by='timepoint')
df.head()
```

Q3. Para o sujeito s1, os sinais apresentam uma variação menor, aproximadamente igual ou maior com relação ao sujeito s0?

```python
s0stim = df[(df.subject == 's0') & (df.region == 'parietal') & (df.event == 'stim')]
s0cue = df[(df.subject == 's0') & (df.region == 'parietal') & (df.event == 'cue')]

s1stim = df[(df.subject == 's1') & (df.region == 'parietal') & (df.event == 'stim')]
s1cue = df[(df.subject == 's1') & (df.region == 'parietal') & (df.event == 'cue')]
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(s0stim.timepoint, s0stim.signal , label='stim')
plt.plot(s0cue.timepoint, s0cue.signal , label='cue')
plt.title('FMRI sujeito S0')
plt.xlabel('tempo')
plt.ylabel('amplitude')
plt.legend()

plt.subplot(1,2,2)
plt.plot(s1stim.timepoint, s1stim.signal , label='stim')
plt.plot(s1cue.timepoint, s1cue.signal , label='cue')
plt.title('FMRI sujeito S1')
plt.xlabel('tempo')
plt.ylabel('amplitude')
plt.legend()

plt.show()
```

# Exercício 6 plt.subplots(nrow,ncol,...) ax[i,<j>]
Produza um gráfico como abaixo para os sujeitos s0 e s1, em uma mesma figura empregando, agora empregando o modo artist layer. Observe todos os demais elementos do gráfico

![imagem](https://github.com/Rogerio-mack/IMT_Ciencia_de_Dados/raw/main/images/ex_matplotlib1_6.png)

Q4. Qual a dimensão do ax criado? (dica: empregue ax.shape)

```python
fig, ax = plt.subplots(1,2,figsize=(12,4)) # cria uma figura alterando o tamanho, figsize=(tamanho_horizontal,tamanho_vertical)

plt.suptitle('Artist Layer', fontsize=14, weight='bold', y=1.05)

# Os comandos podem então ser direcionados para o gráfico específico
ax[0].plot(s0stim.timepoint, s0stim.signal , label='stim')
ax[0].plot(s0cue.timepoint, s0cue.signal , label='cue')
ax[0].set_title('FMRI sujeito S0')
ax[0].set_xlabel('tempo')
ax[0].set_ylabel('amplitude')
ax[0].legend()

ax[1].plot(s1stim.timepoint, s1stim.signal , label='stim')
ax[1].plot(s1cue.timepoint, s1cue.signal , label='cue')
ax[1].set_title('FMRI sujeito S1')
ax[1].set_xlabel('tempo')
ax[1].set_ylabel('amplitude')
ax[1].legend()

plt.tight_layout()
plt.show()

ax.shape
```
